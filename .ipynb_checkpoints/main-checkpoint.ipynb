```python
# Extract features (X) by dropping the 'target' column from the DataFrame
X = df.drop('target', axis=1)

# Extract the target variable (y)
y = df['target']
```

```python
# Set the random seed for reproducibility
np.random.seed(42)

# Split the dataset into training and testing sets
# X_train, X_test: feature matrices for training and testing, respectively
# y_train, y_test: target vectors for training and testing, respectively
# test_size=0.2: 80% of the data will be used for training, and 20% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Display the feature matrix for the training set
X_train
```

```python
# Import Logistic Regression model from scikit-learn
from sklearn.linear_model import LogisticRegression

# Import K-Nearest Neighbors (kNN) Classifier from scikit-learn
from sklearn.neighbors import KNeighborsClassifier

# Import Random Forest Classifier from scikit-learn
from sklearn.ensemble import RandomForestClassifier


# Define a dictionary of models
models = {
    'LogisticRegression': LogisticRegression(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'RandomForestClassifier': RandomForestClassifier(criterion='log_loss')
}

# Set the random seed for reproducibility
np.random.seed(45)

# Create a dictionary to store model scores
models_score = {}

# Train each model and evaluate its performance
for name, model in models.items():
    model.fit(X_train, y_train)
    # Calculate and store the accuracy score on the test set
    models_score[name] = model.score(X_test, y_test)

# Display the model scores
for item in models_score.items():
    print(item)
```

```python
# Create a DataFrame from the models' accuracy scores
model_compare = pd.DataFrame(models_score, index=['accuracy'])

# Transpose the DataFrame and plot a bar chart
model_compare.T.plot.bar()

# Add plot title and labels
plt.title('Model Comparison - Accuracy')
plt.xlabel('Model')
plt.ylabel('Accuracy')

# Display the plot
plt.show()
```

```python
# Set the random seed for reproducibility
np.random.seed(45)

# Create a Logistic Regression model
model_2 = LogisticRegression()

# Train the model on the training data
model_2.fit(X_train, y_train)

# Evaluate the model's accuracy on the test set
accuracy = model_2.score(X_test, y_test)

# Display the accuracy score
print("Accuracy:", accuracy)
```

```python
# Initialize empty lists to store training and test scores
train_scores = []
test_scores = []

# Define a range of values for the number of neighbors (1 to 14)
neighbors = range(1, 15)

# Create a k-Nearest Neighbors (kNN) classifier
knn = KNeighborsClassifier()

# Set the random seed for reproducibility
np.random.seed(45)

# Loop through different values of neighbors and collect training/test scores
for i in neighbors:
    # Set the number of neighbors for the kNN model
    knn.set_params(n_neighbors=i)
    
    # Train the kNN model on the training data
    knn.fit(X_train, y_train)
    
    # Append the training and test scores to the respective lists
    train_scores.append(knn.score(X_train, y_train))
    test_scores.append(knn.score(X_test, y_test))
```

```python
# Find the maximum test accuracy among different values of n_neighbors
best_test_accuracy = max(test_scores)

# Display the maximum test accuracy
print(f"Maximum Test Accuracy: {best_test_accuracy*100:.2f}%")
```

```python
# Plot the training and test scores for different values of n_neighbors
plt.plot(neighbors, train_scores, label='Train Scores')
plt.plot(neighbors, test_scores, label='Test Scores')

# Set x-axis ticks to show values from 1 to 14
plt.xticks(np.arange(1, 15, 1))

# Add legend to differentiate between train and test scores
plt.legend()

# Add plot title and labels
plt.title('Training and Test Scores for Different Values of n_neighbors')
plt.xlabel('Number of Neighbors (n_neighbors)')
plt.ylabel('Accuracy')

# Display the plot
plt.show()
```

```python
# Grid search parameters for Logistic Regression
log_reg_grid = {
    'C': np.logspace(-4, 4, 20),  # Regularization parameter values ranging from 1e-4 to 1e4
    'solver': ['liblinear']       # Solver algorithm for optimization
}

# Grid search parameters for Random Forest Classifier
rf_grid = {
    'n_estimators': np.arange(10, 1000, 50),  # Number of trees in the forest
    'max_depth': [None, 3, 5, 10],             # Maximum depth of the trees
    'min_samples_split': np.arange(2, 20, 2),  # Minimum number of samples required to split an internal node
    'min_samples_leaf': np.arange(1, 20, 2),   # Minimum number of samples required to be at a leaf node
}
```

```python
# Set the random seed for reproducibility
np.random.seed(42)

# Create a RandomizedSearchCV object for Logistic Regression
rs_log_reg = RandomizedSearchCV(
    LogisticRegression(),             # Logistic Regression model to be optimized
    param_distributions=log_reg_grid, # Grid of hyperparameter values for random search
    n_iter=20,                        # Number of iterations for random search
    cv=5,                             # Number of cross-validation folds
    verbose=True                      # Display progress information
)

# Perform the random search on the training data
rs_log_reg.fit(X_train, y_train)
```

```python
# Retrieve the best hyperparameters found during Randomized Search for Logistic Regression
best_log_reg_params = rs_log_reg.best_params_

# Display the best hyperparameters
print("Best Hyperparameters for Logistic Regression:", best_log_reg_params)
```

```python
# Evaluate the accuracy of the Logistic Regression model with the best hyperparameters on the test set
test_accuracy = rs_log_reg.score(X_test, y_test)

# Display the test accuracy
print("Test Accuracy for Logistic Regression:", test_accuracy)
```

```python
# Set the random seed for reproducibility
np.random.seed(42)

# Create a RandomizedSearchCV object for Random Forest Classsifier
rs_rf = RandomizedSearchCV(
    RandomForestClassifier(),          # Random Forest Classsifier model to be optimized
    param_distributions=rf_grid,      # Grid of hyperparameter values for random search
    cv=5,                             # Number of cross-validation folds
    n_iter=10,                        # Number of iterations for random search
    verbose=True                      # Display progress information
)

# Perform the random search on the training data
rs_rf.fit(X_train, y_train)
```

```python
# Retrieve the best hyperparameters found during Randomized Search for Random Forest Classifier
best_rf_params = rs_rf.best_params_

# Display the best hyperparameters
print("Best Hyperparameters for Random Forest Classifier:", best_rf_params)
```

```python
rs_rf.score(X_test, y_test)
```

```python
# Grid search parameters for Logistic Regression
log_reg_grid = {
    'C': np.logspace(-4, 4, 30),  # Regularization parameter values ranging from 1e-4 to 1e4
    'solver': ['liblinear']        # Solver algorithm for optimization
}

# Create a GridSearchCV object for Logistic Regression
gs_log_reg = GridSearchCV(
    LogisticRegression(),             # Logistic Regression model to be optimized
    param_grid=log_reg_grid,          # Grid of hyperparameter values for grid search
    cv=5,                             # Number of cross-validation folds
    verbose=True                      # Display progress information
)

# Perform grid search on Logistic Regression using the training data
gs_log_reg.fit(X_train, y_train)
```

```python
# Retrieve the best hyperparameters found during Grid Search for Logistic Regression
best_log_reg_params = gs_log_reg.best_params_

# Display the best hyperparameters
print("Best Hyperparameters for Logistic Regression:", best_log_reg_params)
```

```python
# Evaluate the accuracy of the Logistic Regression model with the best hyperparameters on the test set
test_accuracy_log_reg = gs_log_reg.score(X_test, y_test)

# Display the test accuracy
print("Test Accuracy for Logistic Regression:", test_accuracy_log_reg)
```

```python
# Generate predictions on the test set using the Logistic Regression model with the best hyperparameters
y_preds = gs_log_reg.predict(X_test)

# Display the predicted values
print("Predicted Values:", y_preds)
```

```python
# Convert the target variable y_test to a NumPy array
y_test_array = np.array(y_test)

# Display the NumPy array
print("NumPy Array for y_test:", y_test_array)
```

```python
import sklearn.metrics as metrics
import matplotlib.pyplot as plt

# Calculate the probability estimates for the positive class
probs = gs_log_reg.predict_proba(X_test)
preds = probs[:, 1]

# Calculate the False Positive Rate (fpr) and True Positive Rate (tpr) for all thresholds
fpr, tpr, threshold = metrics.roc_curve(y_test, preds)

# Calculate the Area Under the Curve (AUC) for the ROC curve
roc_auc = metrics.auc(fpr, tpr)

# Plot the ROC curve
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.plot(fpr, tpr, 'b', label='AUC = %0.2f' % roc_auc)  # Plotting the ROC curve with AUC label
plt.legend(loc='lower right')  # Display the legend in the lower right corner
plt.plot([0, 1], [0, 1], 'r--')  # Diagonal line representing random classification

# Set axis labels and limits
plt.xlim([0, 1])  # x-axis limit from 0 to 1
plt.ylim([0, 1])  # y-axis limit from 0 to 1
plt.ylabel('True Positive Rate (Sensitivity)')  # Label for the y-axis
plt.xlabel('False Positive Rate (1 - Specificity)')  # Label for the x-axis

# Display the plot
plt.show()
```

```python
# Generate a confusion matrix for the true labels (y_test) and predicted labels (y_preds)
conf_matrix = confusion_matrix(y_test, y_preds)

# Display the confusion matrix
print("Confusion Matrix:")
print(conf_matrix)
```

```python
# Import Seaborn
import seaborn as sns
sns.set(font_scale=1.5)  # Increase font size
 
def plot_conf_mat(y_test, y_preds):
    """
    Plots a confusion matrix using Seaborn's heatmap().
    
    Parameters:
    - y_test: True labels.
    - y_preds: Predicted labels.
    """
    # Create a figure and axis for the heatmap
    fig, ax = plt.subplots(figsize=(3, 3))
    
    # Use Seaborn's heatmap to visualize the confusion matrix
    ax = sns.heatmap(confusion_matrix(y_test, y_preds),
                     annot=True,  # Annotate the boxes with values
                     cbar=False)  # Disable color bar
    
    # Set labels for the x and y axes
    plt.xlabel("Predicted label")  # Predictions go on the x-axis
    plt.ylabel("True label")       # True labels go on the y-axis
    
# Call the function with the true labels (y_test) and predicted labels (y_preds)
plot_conf_mat(y_test, y_preds)
```

```python
# Create a new classifier with best parameters
clf = LogisticRegression(C=0.20433597178569418,
                        solver='liblinear')
```

```python
# Perform cross-validation to evaluate the accuracy of the Logistic Regression model
cv_acc = cross_val_score(clf, X, y, scoring='accuracy', cv=5)

# Display the array of accuracy scores for each fold
print("Cross-Validation Accuracy Scores:", cv_acc)
```

```python
# Perform cross-validation to evaluate the precision of the Logistic Regression model
cv_precision = cross_val_score(clf, X, y, scoring='precision', cv=5)

# Display the array of precision scores for each fold
print("Cross-Validation Precision Scores:", cv_precision)
```

```python
# Perform cross-validation to evaluate the recall of the Logistic Regression model
cv_recall = cross_val_score(clf, X, y, scoring='recall', cv=5)

# Display the array of recall scores for each fold
print("Cross-Validation Recall Scores:", cv_recall)
```

```python
# Perform cross-validation to evaluate the F1-score of the Logistic Regression model
cv_f1_score = cross_val_score(clf, X, y, scoring='f1', cv=5)

# Display the array of F1-score scores for each fold
print("Cross-Validation F1-Score Scores:", cv_f1_score)
```

```python
# Calculate the mean accuracy across folds
mean_accuracy = np.mean(cv_acc)
print("Mean Accuracy:", mean_accuracy)

# Calculate the mean precision across folds
mean_precision = np.mean(cv_precision)
print("Mean Precision:", mean_precision)

# Calculate the mean recall across folds
mean_recall = np.mean(cv_recall)
print("Mean Recall:", mean_recall)

# Calculate the mean F1-score across folds
mean_f1_score = np.mean(cv_f1_score)
print("Mean F1-Score:", mean_f1_score)
```

```python
import pandas as pd

# Create a DataFrame to store cross-validation metrics
cv_metrics = pd.DataFrame({
    'Accuracy': cv_acc,
    'Precision': cv_precision,
    'Recall': cv_recall,
    'F1-Score': cv_f1_score
})

# Display the DataFrame
print("Cross-Validation Metrics:")
print(cv_metrics)
```

```python
# Create a bar plot for cross-validation metrics (transpose the DataFrame for better visualization)
ax = cv_metrics.mean().T.plot.bar(rot=0, figsize=(10, 6), legend=False)

# Set plot labels and title
plt.ylabel('Score')
plt.xlabel('Metrics')
plt.title('Cross-Validation Metrics')

# Display the plot
plt.show()
```
